# Convert Text Data to Numerals

```python
docs = [
    'go india',
    'india india',
    'hip hip hurray',
    'jeetega bhai jeetega india jeetega',
    'bharat mata ki jay',
    'kohli kohli',
    'sachin sachin',
    'rohit hitman sharma',
    'modi ji ki jay',
    'inquilab jindabad',
    'pakistan murdabad'
]
```

```python
len(docs)
```

## Tokenize the dictionary
Remove special symbols <br>
Change capital letters to small letters, etc <br>


```python
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
```

Out of vocabulary words are replaced by oov_token val

```python
tokenizer = Tokenizer(oov_token='')
```

```python
tokenizer.fit_on_texts(docs)
```

## Tokenizer functions
tokenizer.word_counts <br>
tokenizer.word_index <br>
tokenizer.word_docs <br>
tokenizer.document_count

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

## Text to Sequences

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

## Apply padding

```python
from keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding='post')
sequences
```
