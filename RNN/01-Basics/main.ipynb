```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
from tensorflow.keras.utils import plot_model
```

```python
model = Sequential(name='input_layer')
model.add(SimpleRNN(units=3, input_shape=(4, 5), name='rnn_layer'))
model.add(Dense(units=1, activation='sigmoid', name='output_layer'))
model.name = 'simple_rnn.keras'
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True, show_layer_names=True, to_file='simple_rnn.png')
```

```python
for layer in model.layers:
    print ("xx"*10)
    print(f"layer name: {layer.name}")
    print(f"layer type: {layer.__class__.__name__}")
    print(f"layer config: {layer.get_config()}")
    for weights in layer.get_weights():
        print("x"*5)
        print(f"weights shape: {weights.shape}")
        print(weights)
    print ("xx"*10)
```
