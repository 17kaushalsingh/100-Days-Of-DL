## Libraries for Data Augmentation

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import load_img, array_to_img, img_to_array, ImageDataGenerator
```

## Read img using image from tf

```python
img = image.load_img('Raw_Data/cat_dog.jpg', target_size=(228, 228))
```

## Augmentation from single image

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

## Convert PIL to numpy image

```python
type(img)

x = img_to_array(img)
```

```python
type(x)
x.shape
```

```python
x = x.reshape((1,) + x.shape)
```

```python
x.shape
```

```python
%pwd
```

```python
cnt = 0

for batch in datagen.flow(x, batch_size=1, save_to_dir='Aug_Data', save_format='jpeg'):
    cnt += 1

    if cnt == 100:
        break
```
