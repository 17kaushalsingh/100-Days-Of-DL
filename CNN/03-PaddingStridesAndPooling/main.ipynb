## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Conv2D
```

## Import Datasets

```python
from tensorflow.keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
plt.imshow(X_train[878])
```

## Create ConvNet

```python
model = Sequential()

## Convolutional Layers
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu'))

## Flatten the layers
model.add(Flatten())

## Fully Connected Layers
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=128, activation='relu'))

## Output Layers
model.add(Dense(units=10, activation='softmax'))

## Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

## ModelSummary
model.name = 'cnn_without_padding_and_strides'
model.summary()

history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test)).history
```

```python
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test), initial_epoch=7).history
```

```python
model.save("model.keras")
model.export("filepath=model.keras", format='onnx')
```

```python
plt.plot(history['loss'], label='train loss', color='red')
plt.plot(history['val_loss'], label='val loss', color='green')
plt.title('Model Loss')
plt.legend()
plt.show()

plt.plot(history['accuracy'], label='train accuracy', color='red')
plt.plot(history['val_accuracy'], label='val accuracy', color='green')
plt.title('Model Accuracy')
plt.legend()
plt.show()
```

```python
y_pred = np.argmax(model.predict(X_test), axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
