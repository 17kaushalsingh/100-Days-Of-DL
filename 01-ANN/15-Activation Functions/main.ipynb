```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy import exp as e
```

# Sigmoid Activation Function

```python
def sigmoid(x):
    f = 1 / (1 + e(-x))
    return f

def d_sigmoid(x):
    f = e(-x) / ((1 + e(-x))**2)
    return f
```

```python
x = np.linspace(-10, 10, 1000)
y = sigmoid(x)
y_dash = d_sigmoid(x)
```

```python
plt.plot(x, y, label='sigmoid fn', color='red')
plt.plot(x, y_dash, label='derivative of sigmoid fn', color='green')
plt.title("SIGMOID ACTIVATION")
plt.xlim(-10,10)
plt.ylim(-0.1, 1.1)
plt.legend()
plt.grid(True)
plt.show()
```

# TanH Activation Function

```python
def tanh(x):
    f = 2 * sigmoid(2*x) - 1
    return f

def d_tanh(x):
    f = 4 * d_sigmoid(2 * x)
    return f
```

```python
x = np.linspace(-8, 8, 10000)
y = tanh(x)
y_dash = d_tanh(x)
```

```python
plt.plot(x, y, label='tanh fn', color='red')
plt.plot(x, y_dash, label='derivative of tanh fn', color='green')
plt.title("TANH ACTIVATION")
plt.legend()
plt.grid(True)
plt.show()
```

# ReLU Activation Function

```python
def relu(x):
    f = np.where(x >= 0, x, 0)
    return f

def d_relu(x):
    f = np.where(x >= 0, 1, 0)
    return f
```

```python
x = np.linspace(-8, 8, 10000)
y = relu(x)
y_dash = d_relu(x)
```

```python
plt.plot(x, y, label='relu fn', color='red')
plt.plot(x, y_dash, label='relu of tanh fn', color='green')
plt.title("RELU ACTIVATION")
plt.legend()
plt.grid(True)
plt.show()
```
