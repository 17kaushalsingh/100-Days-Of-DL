## L1 Loss

```python
x = np.linspace(-20, 20, 300)
l1_loss = np.abs(x)

plt.plot(x, l1_loss, label='l1_loss', color='red')
plt.grid(True)
```

## L2 Loss

```python
x = np.linspace(-20, 20, 300)
l2_loss = x**2

plt.plot(x, l2_loss, label='l2_loss', color='blue')
plt.grid(True)
```

## Huber Loss

```python
x = np.linspace(-20, 20, 300)
delta = 0.9
huber_loss = np.where(x <= delta, (x**2)/2.0, delta*np.abs(x) - (delta**2)/2)

plt.plot(x, huber_loss, label='huber_loss', color='green')
plt.grid(True)
```

# Comparison

```python
x = np.linspace(-2, 2, 300)

l1_loss = np.abs(x)

l2_loss = x**2

delta = 0.5
huber_loss = np.where(x <= delta, (x**2)/2.0, delta*np.abs(x) - (delta**2)/2)

plt.plot(x, l1_loss, label='l1_loss', color='red')
# plt.plot(x, l2_loss, label='l2_loss', color='blue')
plt.plot(x, huber_loss, label='huber_loss', color='green')
plt.grid(True)
plt.legend()
plt.title('Regression Losses')
```

# Logistic Loss

```python
x = np.linspace(-100, 100, 100000)
log_loss = -x*np.log(x) - (1-x)*np.log(1-x)

plt.plot(x, log_loss, label='log_loss', color='green')
plt.grid(True)
```
