# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import tensorflow
from tensorflow import keras

## For Sequential
from tensorflow.keras import Sequential
from tensorflow.keras.layers import *

## For Functional
from tensorflow.keras import Model
```

# Functional API Model
### We must define input shape and output shape for functional API

```python
## Input Shape/Layer
x = Input(shape=(3,))
```

```python
## Common Layers
hidden_1 = Dense(units=128, activation='relu', name='hidden_layer_1')(x) ## Hidden Layer 1
hidden_2 = Dense(units=64, activation='relu', name='hidden_layer_2')(hidden_1) ## Hidden Layer 2
hidden_3 = Dense(units=32, activation='relu', name='hidden_layer_3')(hidden_2) ## Hidden Layer 3

## 2 Output Layers
output_1 = Dense(units=1, activation='linear', name='output_layer_age')(hidden_3) ## Output Layer 1
output_2 = Dense(units=1, activation='sigmoid', name='output_layer_state')(hidden_3) ## Output Layer 2
```

```python
## Define Model
model = Model(inputs=x, outputs=[output_1, output_2])
```

```python
model.name = 'Functional Model'
model.summary()
```

# Visualize Model

```python
from keras.utils import plot_model

plot_model(model, show_shapes=True, show_layer_names=True, to_file='model_problem1.png')
```
